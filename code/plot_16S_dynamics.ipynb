```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read 16S data

```python
df_16S = pd.read_excel('../data/16S_absolute_abundance.xlsx', index_col=0)
df_16S.head()
```

# Read meta data

```python
df_meta = pd.read_excel('../data/meta_data.xlsx', index_col=0)
df_meta.head()
```

# Combine data tables

```python
df_16S_meta = pd.merge(df_meta, df_16S, left_index=True, right_index=True, how='inner')
df_16S_meta.head()
```

# Plot

```python
taxa_to_plot = ['Muribaculaceae',
               'Bacteroides-acidifaciens',
               'Faecalibaculum',
               'Bacteroides',
               'Parasutterella',
               'Lachnospiraceae-bacterium-28-4',
               'Parabacteroides-goldsteinii',
               'Desulfovibrionaceae']
```

```python
fig, ax = plt.subplots(figsize=(14,16), nrows=len(taxa_to_plot), ncols=4, sharex=True, sharey=False)
Vendor = ['Shanghai','Hunan','Guangdong','Beijing']
for idx_i, b in enumerate(taxa_to_plot):
    for idx_j, v in enumerate(Vendor):
        curr_df = df_16S_meta[(df_16S_meta.Vendor==v) & (df_16S_meta.Diet=='Inulin')]
        sns.lineplot(x='Day',y=b, marker='o', ci=68, data=curr_df, ax=ax[idx_i,idx_j], color='green') # ci=68 correspond to standard error
        if idx_i==0:
            ax[idx_i,idx_j].set_title(v)
            
        max_height = curr_df[b].max()
        exponent = np.floor(np.log10(max_height))
        leading_digit = int(max_height/10**exponent)+1
        max_height = leading_digit*10**exponent
        ax[idx_i,idx_j].set_ylim([-max_height/10,max_height])
        ax[idx_i,idx_j].set_yticks([0,max_height/2,max_height])
        if idx_j!=0:
            ax[idx_i,idx_j].set_ylabel('')
    
        if idx_i==0:
            ax[idx_i,idx_j].set_ylim([-0.4e8,4e8])
            ax[idx_i,idx_j].set_yticks([0,2e8,4e8])
        if idx_i==1:
            ax[idx_i,idx_j].set_ylim([-0.3e8,3e8])
            ax[idx_i,idx_j].set_yticks([0,1.5e8,3e8])
        if idx_i==2:
            ax[idx_i,idx_j].set_ylim([-0.8e7,8e7])
            ax[idx_i,idx_j].set_yticks([0,4e7,8e7])
        if idx_i==3:
            ax[idx_i,idx_j].set_ylim([-0.6e7,6e7])
            ax[idx_i,idx_j].set_yticks([0,3e7,6e7])
        if idx_i==4:
            ax[idx_i,idx_j].set_ylim([-0.5e7,5e7])
            ax[idx_i,idx_j].set_yticks([0,2.5e7,5e7])
        if idx_i==5:
            ax[idx_i,idx_j].set_ylim([-0.2e7,2e7])
            ax[idx_i,idx_j].set_yticks([0,1e7,1e7])
        if idx_i==6:
            ax[idx_i,idx_j].set_ylim([-0.3e7,3e7])
            ax[idx_i,idx_j].set_yticks([0,1.5e7,1.5e7])
```

```python
plt.rcParams['svg.fonttype'] = 'none'
plt.savefig('taxa_16S_dynamics_lineplot.svg', format='svg')
```
