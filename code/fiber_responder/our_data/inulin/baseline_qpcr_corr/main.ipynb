```python
import pandas as pd
import numpy as np
import scipy
from copy import deepcopy
import matplotlib.pyplot as plt
from sklearn.metrics import auc
import seaborn as sns
from scipy import stats
from scipy.stats import spearmanr, pearsonr
import warnings
import itertools
warnings.filterwarnings("ignore")
```

# Read 16S data

```python
df_16S_relab = pd.read_excel('../../../../../our_data/16S_relative_abundance.xlsx', index_col=0)
df_16S_relab = df_16S_relab[['Bacteroides-acidifaciens','Muribaculaceae','Faecalibaculum','Parasutterella','Bacteroides']]
df_16S_relab.columns = [x+'__relab' for x in df_16S_relab.columns]
df_16S_relab.head()
```

# Read qPCR data

```python
df_qpcr = pd.read_excel('../../../../../our_data/qPCR.xlsx', index_col=0)
df_qpcr.columns= ['qPCR']
df_qpcr.head()
```

# Read metadata for inulin

```python
df_meta = pd.read_excel('../../../../../our_data/meta_data.xlsx', index_col=0)
df_meta = df_meta[df_meta.Diet=='Inulin']
df_16S_relab = df_16S_relab.loc[df_meta.index]
df_qpcr = df_qpcr.loc[df_meta.index]
df_meta = pd.merge(df_meta, df_16S_relab, left_index=True, right_index=True)
df_meta = pd.merge(df_meta, df_qpcr, left_index=True, right_index=True)

# focus on early phasae (5 day)
df_meta = df_meta[df_meta.Day<=5]

df_meta.head()
```

# Calculate area under the curve

```python
lines = []
inulin_responders = ['Bacteroides-acidifaciens','Muribaculaceae','Faecalibaculum','Parasutterella','Bacteroides']
for mice in set(df_meta.MiceID):
    curr_df = df_meta[df_meta.MiceID==mice].sort_values(by='Day')
    vendor = list(set(curr_df['Vendor']))[0]
    assert np.max(curr_df.Day)==5
    for responder in inulin_responders:
        baseline = list(curr_df.loc[curr_df.Day==0,responder+'__relab'])[0]
        lines.append([mice, vendor, responder, baseline, auc(list(curr_df.Day), list(curr_df.qPCR))])
df_auc = pd.DataFrame(lines, columns=['MiceID','Vendor','Responder','Baseline','AUC'])
df_auc.head()
```

# Correlation

```python
lines = []
for responder in inulin_responders:
    curr_df = df_auc[df_auc.Responder==responder]
    lines.append([responder, pearsonr(curr_df.Baseline, curr_df.AUC)[0], spearmanr(curr_df.Baseline, curr_df.AUC)[0]])
df_corr = pd.DataFrame(lines, columns=['Responder','Pearson','Spearman'])
df_corr.head()
```

```python
fig, ax=plt.subplots(figsize=(6,4))
_ = sns.barplot(x='Responder', y='Spearman', data=df_corr, ax=ax)
plt.savefig('spearman_corr_baseline_qpcr_auc.svg', format='svg')
```
