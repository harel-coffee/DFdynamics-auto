# Plot SCFA producer and degrader

```python
fig, ax = plt.subplots(figsize=(15,8), nrows=2, ncols=3, sharex=True, sharey=True)
curr_df = df_16S_meta[df_16S_meta.Diet=='Inulin']

# acetate producer
ac_producer = ['Parabacteroides-goldsteinii','Lachnospiraceae-bacterium-28-4']
for t in ac_producer:
    sns.lineplot(x='Day',y=t, marker='o', ci=68, data=curr_df, ax=ax[0,0])
ax[0,0].set_title('Acetate')
ax[0,0].set_ylabel('SCFA producer')

# acetate degrader
ac_degrader = ['Clostridia-vadinBB60-group','Clostridia-UCG-014']
for t in ac_degrader:
    sns.lineplot(x='Day',y=t, marker='o', ci=68, data=curr_df, ax=ax[1,0])
ax[1,0].set_ylabel('SCFA degrader')

# butyrate producer
but_producer = ['Lachnospiraceae-bacterium-28-4']
for t in but_producer:
    sns.lineplot(x='Day',y=t, marker='o', ci=68, data=curr_df, ax=ax[0,1])
ax[0,1].set_title('Butyrate')

# butyrate degrader
but_degrader = ['Clostridia-UCG-014']
for t in but_degrader:
    sns.lineplot(x='Day',y=t, marker='o', ci=68, data=curr_df, ax=ax[1,1])
ax[1,1].set_ylabel('SCFA degrader')

# propionate producer
but_producer = ['Lachnospiraceae-bacterium-28-4']
for t in but_producer:
    sns.lineplot(x='Day',y=t, marker='o', ci=68, data=curr_df, ax=ax[0,1])
ax[0,1].set_title('Butyrate')

# propionate degrader
but_degrader = ['Clostridia-UCG-014']
for t in but_degrader:
    sns.lineplot(x='Day',y=t, marker='o', ci=68, data=curr_df, ax=ax[1,1])
ax[1,1].set_ylabel('SCFA degrader')


```
