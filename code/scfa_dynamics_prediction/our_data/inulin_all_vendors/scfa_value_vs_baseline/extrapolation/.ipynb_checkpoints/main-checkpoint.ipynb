# Get data

```python
df_bac = pd.read_csv('../../../../../../our_data/16S_relative_abundance_species.csv', index_col=0)
df_meta_tmp = pd.read_csv('../../../../../../our_data/meta_data.csv', index_col=0)
df_meta_tmp = df_meta_tmp[(df_meta_tmp.Day==0) & (df_meta_tmp.Diet=='Inulin')]
df_bac = df_bac.loc[df_meta_tmp.index]
df_bac = df_bac.rename(dict(zip(df_meta_tmp.index,df_meta_tmp.MiceID)))

# standardization
scaler = StandardScaler()
scaler.fit(df_bac.T.values)
df_bac = pd.DataFrame(scaler.transform(df_bac.T.values), index=df_bac.columns, columns=df_bac.index).T

df_bac.head()
```

```python
df_bfit = pd.read_excel('../../../../../harmonic_fitting/inulin/best_fit_parameters_r2_fixedIC.xlsx', index_col=0)
df_meta = pd.read_csv('../../../../../../our_data/meta_data.csv', index_col=0)
df_meta = df_meta[df_meta.Diet=='Inulin'].reset_index(drop=True).drop(['Day','Diet'],axis=1).drop_duplicates().reset_index(drop=True)
df_meta = pd.merge(df_bfit, df_meta, left_on=['MiceID','Vendor'], right_on=['MiceID','Vendor'], how='outer')
df_meta.head()
```
