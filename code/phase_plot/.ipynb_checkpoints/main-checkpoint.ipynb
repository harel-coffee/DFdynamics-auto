```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.metrics import r2_score
from scipy.spatial import distance
from skbio.stats.ordination import pcoa
%run -i '../../utils.py'
```

# Get data

```python
# read meta data
df_meta = pd.read_excel('../../our_data/meta_data.xlsx', index_col=0)
df_meta = df_meta[df_meta.Diet=='Inulin']
df_meta = df_meta.drop('Diet', axis=1)
df_meta = df_meta.rename({'MiceID':'SubjectID'}, axis=1)
df_meta.head()
```

```python
# read SCFA data
df_scfa = pd.read_excel('../../our_data/SCFA.xlsx', index_col=0)
df_scfa.head()
```

```python
# read bacterial abundance
#df_bac = pd.read_excel('../../our_data/16S_absolute_abundance.xlsx', index_col=0)

# run PcoA
dist = distance.squareform(distance.pdist(df_bac, metric="braycurtis"))
df_dist = pd.DataFrame(dist, index = df_bac.index, columns = df_bac.index)
OrdinationResults = pcoa(df_dist.values, number_of_dimensions=3)
df_ord = pd.DataFrame(OrdinationResults.samples.values, index=df_bac.index, columns=['PC1','PC2','PC3'])
df_ord.head()
```

```python
df_join = pd.merge(df_meta, df_scfa, left_index=True, right_index=True, how='inner')
df_join = pd.merge(df_join, df_ord, left_index=True, right_index=True, how='inner')
df_join.head()
```
