# Propionate

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.spatial.distance import pdist
import numpy as np
from skbio import DistanceMatrix
from skbio.stats.distance import permanova

df_meta = pd.read_csv('../../our_data/meta_data.csv',index_col=0)
df_scfa = pd.read_csv('../../our_data/SCFA.csv',index_col=0)
df = pd.merge(df_meta, df_scfa, left_index=True, right_index=True, how='inner')

fig, ax=plt.subplots(nrows=1,ncols=2,figsize=(10,4),sharey=True,sharex=True)
PROPS = {
    'boxprops':{'facecolor':'none', 'edgecolor':'black'},
    'medianprops':{'color':'black'},
}
delta_propionate = []
for index in df.index:
    miceid = df.loc[index,'MiceID']
    baseline = list(df.loc[(df.MiceID==miceid)&(df.Day==0),'Propionate'])[0]
    delta_propionate.append(df.loc[index,'Propionate']-baseline)
df['delta_propionate'] = delta_propionate

# day 5
df_day5 = df[(df.Diet=='Inulin')&(df.Day==5)]
_ = sns.boxplot(x='Vendor', y='delta_propionate', data=df_day5, ax=ax[0], showfliers=False, **PROPS)
_ = sns.stripplot(x='Vendor', y='delta_propionate', data=df_day5, ax=ax[0], jitter=True, size=10)
dm = pdist(np.asarray(df_day5['delta_propionate'].values).reshape(-1,1), metric='minkowski')
DM = DistanceMatrix(dm,list(df_day5.MiceID))
stats = permanova(DM, grouping=list(df_day5.Vendor), permutations=999)
_ = ax[0].set_title('Day 5, P = %2.3f'%(stats['p-value']))

# day 31
df_day31 = df[(df.Diet=='Inulin')&(df.Day==31)]
_ = sns.boxplot(x='Vendor', y='delta_propionate', data=df_day31, ax=ax[1], showfliers=False, **PROPS)
_ = sns.stripplot(x='Vendor', y='delta_propionate', data=df_day31, ax=ax[1], jitter=True, size=10)
dm = pdist(np.asarray(df_day31['delta_propionate'].values).reshape(-1,1), metric='minkowski')
DM = DistanceMatrix(dm,list(df_day31.MiceID))
stats = permanova(DM, grouping=list(df_day5.Vendor), permutations=999)
_ = ax[1].set_title('Day 31, P = %2.3f'%(stats['p-value']))

plt.tight_layout()
plt.rcParams['svg.fonttype'] = 'none'
fig.savefig("propionate_day5_day31.svg", format="svg")
```
