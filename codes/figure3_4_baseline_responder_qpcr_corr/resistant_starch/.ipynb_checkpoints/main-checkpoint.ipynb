```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import spearmanr, pearsonr
from scipy.integrate import simps
import warnings
warnings.filterwarnings("ignore")
```

# Read data

```python
# read bacterial relative abundance
df_16S_relab = pd.read_csv('../../../our_data/16S_relative_abundance_species.csv', index_col=0)

# read qPCR
df_qpcr = pd.read_csv('../../../our_data/qPCR.csv', index_col=0).rename({'Value':'qPCR'},axis=1)

# read meta data
df_meta = pd.read_csv('../../../our_data/meta_data.csv', index_col=0)
df_meta = df_meta[df_meta.Diet=='Resistant starch']
df_16S_relab = df_16S_relab.loc[df_meta.index]
df_qpcr = df_qpcr.loc[df_meta.index]
df_meta = pd.merge(df_meta, df_16S_relab, left_index=True, right_index=True)
df_meta = pd.merge(df_meta, df_qpcr, left_index=True, right_index=True)

# calculate total relative abundance of rs responders
rs_responders = ['Muribaculaceae','Faecalibaculum','Desulfovibrionaceae']
df_meta['AllResponders'] = df_meta[rs_responders].sum(axis=1)
df_meta['MuDe'] = df_meta[['Muribaculaceae','Desulfovibrionaceae']].sum(axis=1)
df_meta['MuFa'] = df_meta[['Muribaculaceae','Faecalibaculum']].sum(axis=1)
df_meta['FaDe'] = df_meta[['Faecalibaculum','Desulfovibrionaceae']].sum(axis=1)

df_meta.head()
```

# Predict absolute abundance from relative abundance

```python
lines = []
covariants = ['Muribaculaceae','Faecalibaculum','Desulfovibrionaceae','MuDe','AllResponders']
for responder in covariants:
    pearson_coef, pearson_P = pearsonr(np.asarray(df_meta[responder]), np.asarray(df_meta.qPCR))
    spearman_coef, spearman_P = spearmanr(np.asarray(df_meta[responder]), np.asarray(df_meta.qPCR))
    lines.append([responder, pearson_coef, pearson_P, spearman_coef, spearman_P, pearson_coef**2])
df_responder_qpcr_assoc = pd.DataFrame(lines, columns=['RSResponder','Pearson','Pearson_P','Spearman','Spearman_P','R2'])
df_responder_qpcr_assoc
```

```python
fig, ax=plt.subplots(nrows=1, ncols=2, figsize=(10,4))

# Pearson
_ = sns.barplot(x='RSResponder', y='Pearson', data=df_responder_qpcr_assoc, ax=ax[0], ci=68)

# Bacteroides-acidifaciens + Muribaculaceae
_ = sns.regplot(x='Muribaculaceae', y='qPCR', color='gray', data=df_meta, ax=ax[1], ci=68)
_ = sns.scatterplot(x='Muribaculaceae', y='qPCR', hue='Vendor', size='Day', data=df_meta, ax=ax[1])
_ = ax[1].set_xlim([-0.08,0.8])

plt.rcParams['svg.fonttype'] = 'none'
plt.savefig('correlation_rs_responder_vs_qpcr.svg', format='svg')
```

# Predict response (averaged absolute abundance) from total relative abundance of rs responders in the baseline

```python
lines = []
covariants = ['Muribaculaceae','Faecalibaculum','Desulfovibrionaceae', 'Mu''AllResponders']

lines = []
lines2 = []
for responder in covariants:
    initial_relab = []
    qpcr_ave = []
    for mice in set(df_meta.MiceID):
        df_tmp = df_meta[df_meta.MiceID==mice].sort_values(by='Day')
        vendor = list(set(df_tmp.Vendor))[0]
        x = list(df_tmp.Day)
        y = list(df_tmp.qPCR)
        initial_relab.append(df_tmp.loc[df_tmp.Day==0,responder].values[0])
        qpcr_ave.append(simps(y,x))#/np.max(x))
        
        if responder=='BaMu':
            lines2.append([initial_relab[-1],qpcr_ave[-1],vendor])
    pearson_coef, pearson_P = pearsonr(initial_relab, qpcr_ave)
    spearman_coef, spearman_P = spearmanr(initial_relab, qpcr_ave)
    lines.append([responder, pearson_coef, pearson_P, spearman_coef, spearman_P, pearson_coef**2])
df_responder_qpcr_assoc = pd.DataFrame(lines, columns=['rsResponder','Pearson','Pearson_P','Spearman','Spearman_P','R2'])
df_responder_qpcr_assoc
```
