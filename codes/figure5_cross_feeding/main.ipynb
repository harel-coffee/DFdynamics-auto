```python
import pandas as pd
import numpy as np
import cobra
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import os
%run -i 'utils.py'
%run -i 'build_cobra_model.py'
```

# Models to include

```python
models = {}
for filename in os.listdir('./'):
    if filename.endswith(".xml"):
        print(filename)
        model = cobra.io.read_sbml_model(filename)
        model.solver = 'cplex'
        models[filename.replace('.xml','')] = model
```

# Essential nutrients

```python
# essential nutrients of Roseburia
# zn2, mg2, mn2, mqn7, pheme, cl, cobalt2, succ, fe2, fe3, ca2, so4, fol, cu2, k

essential_nutrients = []
for name, curr_model in models.items():
    curr_model_copy = deepcopy(curr_model)
    for rxn in curr_model_copy.exchanges:
        rxn.lower_bound = -1000
    res = cobra.flux_analysis.single_reaction_deletion(curr_model_copy, curr_model_copy.exchanges)
    res = res[res['growth'] == 0.0].set_index('ids')
    for k in res.index:       
        essential_nutrients.append([name, curr_model_copy.metabolites.get_by_id(list(k)[0].replace('EX_','').replace('(e)','[e]')).name])
df_essential_nutrients = pd.DataFrame(essential_nutrients, columns=['model','nutrient'])
df_essential_nutrients['essential'] = [int(1) for index in df_essential_nutrients.index]
df_essential_nutrients = pd.pivot_table(df_essential_nutrients, values='essential', index='model', columns='nutrient').fillna(int(0))
df_essential_nutrients.head()
```

```python
ax = sns.clustermap(df_essential_nutrients,
                    method='average',
                    metric='cityblock',
                    figsize=(8,4),
                    cmap=matplotlib.colors.ListedColormap(['white', 'red']),
                    vmin=0, 
                    vmax=1,
                    linewidths=0.5,
                    linecolor='black'
                   )

for tick_label in ax.ax_heatmap.axes.get_yticklabels():
    tick_text = tick_label.get_text()
    if tick_text.startswith('Bacteroides'):
        tick_label.set_color('royalblue')
    else:
        tick_label.set_color('forestgreen')
```

# Simulate minimum and maximum SCFA production (western diet)

```python
# load diet flux
dietflux = pd.read_csv('WesternDietAGORA.txt', sep='\t', header=None, index_col=0)
dietflux = dietflux.rename({k:k.replace('(e)','').replace('EX_','') for k in dietflux.index})

# load enzyme cost profiles
enzyme_cost_profile = pd.read_csv('enzyme_cost_agora_genus.csv', index_col=0)

# build metabolic model and simulate
lines = []
for model in ['Bifidobacterium_1678.xml','Clostridium_1485.xml','Roseburia_841.xml']:
    com = build_community_model(
        taxonomy={model:0.5, model.replace('.xml','')+'_clone.xml':0.5},
        AGORA_path='./',
        model_id='com_Bacteroides_816__%s'%(model.replace('.xml','')),
        medium=dietflux,
        enzyme_cost_profile=enzyme_cost_profile,
        biomass_coupling=True
    )
#     com = build_community_model(
#         taxonomy={'Bacteroides_816.xml':0.5, model:0.5},
#         AGORA_path='./',
#         model_id='com_Bacteroides_816__%s'%(model.replace('.xml','')),
#         medium=dietflux,
#         enzyme_cost_profile=enzyme_cost_profile,
#         biomass_coupling=True
#     )
    max_growth = com.slim_optimize()
    print('max growth rate = %2.3f'%(max_growth))

    # simulate metabolic potential
    fraction_of_optimum = np.linspace(0,1,101)
    for f in fraction_of_optimum:
        com2 = deepcopy(com)
        fva_old_objective = com2.problem.Variable("fva_old_objective", lb=f*max_growth, ub=f*max_growth)
        fva_old_obj_constraint = com2.problem.Constraint(com2.solver.objective.expression - fva_old_objective, lb=0, ub=0, name="fva_old_objective_constraint")
        com2.add_cons_vars([fva_old_objective, fva_old_obj_constraint])

        com2.objective = Zero  # This will trigger the reset as well
        ex_rxn_id = 'EX_M_%s_m'%('but')
        if ex_rxn_id in com2.reactions:
            ex_rxn = com2.reactions.get_by_id(ex_rxn_id)
            com2.solver.objective.set_linear_coefficients({ex_rxn.forward_variable: 1, ex_rxn.reverse_variable: -1})

            # maximization
            com2.solver.objective.direction = 'max'
            sol_max = com2.optimize()
            status_max = com2.solver.problem.solution.get_status()
            max_secr_flux = sol_max.objective_value
            print(f, 'max', max_secr_flux, status_max)

            # minimization
            com2.solver.objective.direction = 'min'
            sol_min = com2.optimize()
            status_min = com2.solver.problem.solution.get_status()
            min_secr_flux = sol_min.objective_value
            print(f, 'min', min_secr_flux, status_min)

            lines.append([model.replace('.xml',''), max_growth, f, max_secr_flux, status_max, min_secr_flux, status_min])
            com2.solver.objective.set_linear_coefficients({ex_rxn.forward_variable: 0, ex_rxn.reverse_variable: 0})
        else:
            # minimum CPLEX solution status is 1, here we use 0 to indicate that exchange reaction is not found in metabolic model and no optimization was run
            lines.append([model.replace('.xml',''), max_growth, f, 0.0, 0, 0.0, 0])

df_res = pd.DataFrame(lines, columns=['Genus','MaxGrowth','FracOpt','MaxSecrFlux','MaxCplexStatus','MinSecrFlux','MinCplexStatus'])
if len(df_res.index) == 0:
    print('empty table at %s'%(sampleid))
    assert 0
else:
    df_res.to_csv('butyrate_secretion_flux_western_diet_clone.csv')
```

# Plot

```python
df = None
for diet in ['mouse','western','highfiber']:
    # interspecific
    df_tmp1 = pd.read_csv('butyrate_secretion_flux_%s_diet.csv'%(diet), index_col=0)
    df_tmp1['Diet'] = diet
    df_tmp1['Interspecific'] = True
    df_tmp1['Growth'] = df_tmp1['MaxGrowth']*df_tmp1['FracOpt']
    # intraspecific
    df_tmp2 = pd.read_csv('butyrate_secretion_flux_%s_diet_clone.csv'%(diet), index_col=0)
    df_tmp2['Diet'] = diet
    df_tmp2['Interspecific'] = False
    df_tmp2['Growth'] = df_tmp2['MaxGrowth']*df_tmp2['FracOpt']

    if df is None:
        df = df_tmp1
        df = pd.concat([df, df_tmp2], ignore_index=False)
    else:
        df = pd.concat([df, df_tmp1, df_tmp2], ignore_index=False)

fig, ax = plt.subplots(nrows=3, ncols=3, figsize=(8,8), sharey=True, sharex=True)
for k1,diet in enumerate(['mouse','western','highfiber']):
    for k2,genus in enumerate(['Bifidobacterium_1678','Clostridium_1485','Roseburia_841']):
        df_tmp = df[(df.Diet==diet) & (df.Genus==genus)]
        _ = sns.lineplot(x='Growth', y='MaxSecrFlux', hue='Interspecific', data=df_tmp, marker=None, ax=ax[k1,k2])
        _ = ax[k1,k2].set_title(genus)
        _ = ax[k1,k2].set_ylabel(diet)
        _ = ax[k1,k2].set_xlim([-0.13,1.3])
        _ = ax[k1,k2].set_ylim([-2.5,25])

plt.rcParams['svg.fonttype'] = 'none'
fig.savefig("cross_feeding_analysis.svg", format="svg")
```
