```python
import pandas as pd
import numpy as np
from copy import deepcopy
from scipy.interpolate import CubicSpline
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import matplotlib
from scipy.stats import norm
from scipy.integrate import solve_ivp
import warnings
import seaborn as sns
warnings.filterwarnings("ignore")
import arviz as az
import scipy.stats.kde as kde
```

# Generate files for stan program

```python
# read metadata
df_meta = pd.read_excel('2019_Patnode_cell_all.xlsx', dtype={'MouseID':str})
df_meta['Diet'] = [x.split('+')[1].strip() if '+' in x else 'Control' for x in df_meta['Diet']]
df_meta['MouseID'] = df_meta['MouseID'].astype(str)
df_meta['MouseID'] = [x+':'+y+':'+z for x,y,z in zip(df_meta['Microbial community'],df_meta['MouseID'], df_meta['Diet'])]
bacteria = ['Bacteroides_ovatus_ATCC_8483', 'Bacteroides_cellulosilyticus_WH2','Bacteroides_thetaiotaomicron_7330',
            'Bacteroides_thetaiotaomicron_VPI_5482','Bacteroides_vulgatus_ATCC_8482', 'Bacteroides_caccae',
            'Bacteroides_finegoldii', 'Bacteroides_massiliensis','Collinsella_aerofaciens', 'Escherichia_coli',
            'Odoribacter_splanchnicus', 'Parabacteroides_distasonis','Ruminococcaceae', 'Ruminococcus_albus', 'Subdoligranulum_variabile']
for b in bacteria:
    df_meta[b] = df_meta[b]*df_meta['Fecal DNA Yield (ug/mg)']
df_meta = df_meta.drop(['Microbial community','Fecal DNA Yield (ug/mg)'], axis=1)

for diet in ['Citrus Pectin', 'Orange Peel', 'Pea Fiber', 'Tomato Peel']:
    df_meta2 = df_meta[(df_meta.Diet==diet) | (df_meta.Diet=='Control')]
    df_meta2['Diet'] = [1 if g==diet else 0 for g in df_meta2['Diet']]
    
    # for repeated measurements, take the average
    df_meta2 = df_meta2.groupby(['Day Post Gavage','Diet','MouseID']).agg(np.mean).reset_index()

    # replace zeros with minimum non-zero values (since these values will be log-tranformed later)
    for b in bacteria:
        minval = np.min(df_meta2.iloc[np.nonzero(df_meta2[b].values)[0]][b]) # minimum non-zero value
        df_meta2.loc[df_meta2[b]==0, b] = minval

    # normalize max abosolute value to 1
    normalization_factor = df_meta2[bacteria].max().max()
    df_meta2[bacteria] = df_meta2[bacteria]/normalization_factor
    
    # sort days
    df_meta2 = df_meta2.sort_values(['Diet','MouseID','Day Post Gavage'])

    # calculate derivative of bacterial abundance
    df_deriv = deepcopy(df_meta2)
    for curr_mice in set(df_deriv.MouseID):
        curr_df = df_deriv[df_deriv.MouseID==curr_mice]
        for b in bacteria:
            xdata = np.array(curr_df['Day Post Gavage'])
            ydata = np.array(curr_df[b])
            cs = CubicSpline(xdata, ydata)
            csd1 = cs.derivative(nu=1)
            ydata_d1 = csd1(xdata)
            df_deriv.loc[df_deriv.MouseID==curr_mice, b] = ydata_d1

    # construct regression matrix (Y~X)
    Ymat = df_deriv[bacteria].values # Ymat should be samples by bacteria
    Ymat = Ymat.flatten(order='F')
    Ymat = StandardScaler().fit_transform(Ymat.reshape(-1,1)).reshape(1,-1)[0] # standardize

    Xmat = np.zeros(shape=(len(bacteria)*len(df_deriv.index), (len(bacteria)+2)*len(bacteria)))
    for k in np.arange(len(bacteria)):
        Xmat[k*len(df_deriv.index):(k+1)*len(df_deriv.index),k*(len(bacteria)+2)] = 1
        Xmat[k*len(df_deriv.index):(k+1)*len(df_deriv.index),k*(len(bacteria)+2)+1] = df_deriv.Diet.values
        Xmat[k*len(df_deriv.index):(k+1)*len(df_deriv.index),k*(len(bacteria)+2)+2:(k+1)*(len(bacteria)+2)] = df_meta2[bacteria].values

    # write data file
    json_str = '{\n"N" : %d,\n'%(len(Ymat))
    json_str += '\"dlogX\" : [%s],\n'%(','.join(list(Ymat.astype(str)))) # y variable
    for k1,c1 in enumerate(bacteria): # x variable
        # growth rate
        json_str += '\"growth_rate_%s\" : [%s],\n'%(c1,','.join(list(Xmat[:,k1*(len(bacteria)+2)].astype(str))))
        # diet response
        json_str += '\"diet_response_%s\" : [%s],\n'%(c1,','.join(list(Xmat[:,k1*(len(bacteria)+2)+1].astype(str))))
        # bacterial interactions
        for k2,c2 in enumerate(bacteria):
            v = list(Xmat[:,k1*(len(bacteria)+2)+2+k2].astype(str))
            json_str += '\"pairwise_interaction_%s_%s\" : [%s]'%(c1,c2,','.join(v))
            if c1 == bacteria[-1] and c2 == bacteria[-1]:
                json_str += '\n}'
            else:
                json_str += ',\n'

    text_file = open("%s.data.json"%(diet), "w")
    text_file.write("%s" % json_str)
    text_file.close()

    # write model file
    # data block
    model_str = 'data {\n'
    model_str += '\tint<lower=0> N;\n'
    model_str += '\tvector[N] dlogX;\n'
    for c1 in bacteria:
        model_str += '\tvector[N] growth_rate_%s;\n'%(c1)
        model_str += '\tvector[N] diet_response_%s;\n'%(c1)
        for c2 in bacteria:
            model_str += '\tvector[N] pairwise_interaction_%s_%s;\n'%(c1,c2)
    model_str += '}\n'

    # parameter block
    model_str += 'parameters {\n\treal<lower=0,upper=1> sigma;\n'
    for c1 in bacteria:
        model_str += '\treal alpha__%s;\n'%(c1) # growth rate
        model_str += '\treal epsilon__%s;\n'%(c1) # inulin response
        for c2 in bacteria:
            model_str += '\treal beta__%s_%s;\n'%(c1,c2)
    model_str += '}\n'       

    # model block
    model_str += 'model {\n\tsigma ~ uniform(0,1);\n'
    for c1 in bacteria:
        model_str += '\talpha__%s ~ normal(0,1);\n'%(c1) # growth rate
        model_str += '\tepsilon__%s ~ normal(0,1);\n'%(c1) # inulin response
        for c2 in bacteria:
            model_str += '\tbeta__%s_%s ~ normal(0,1);\n'%(c1,c2)
    model_str += '\tdlogX ~ normal('
    for c1 in bacteria:
        model_str += 'alpha__%s*growth_rate_%s+'%(c1,c1) # growth rate
        model_str += 'epsilon__%s*diet_response_%s+'%(c1,c1) # inulin response
        for c2 in bacteria:
            if c1 == bacteria[-1] and c2 == bacteria[-1]:
                model_str += 'beta__%s_%s*pairwise_interaction_%s_%s'%(c1,c2,c1,c2)
            else:
                model_str += 'beta__%s_%s*pairwise_interaction_%s_%s+'%(c1,c2,c1,c2)
    model_str += ', sigma);\n}'

    text_file = open("%s.stan"%(diet), "w")
    text_file.write("%s" % model_str)
    text_file.close()
```

# Read stan output and generate summary table

```python
def hpd_grid(sample, alpha=0.05, roundto=2):
    """Calculate highest posterior density (HPD) of array for given alpha. 
    The HPD is the minimum width Bayesian credible interval (BCI). 
    The function works for multimodal distributions, returning more than one mode
    Parameters
    ----------
    
    sample : Numpy array or python list
        An array containing MCMC samples
    alpha : float
        Desired probability of type I error (defaults to 0.05)
    roundto: integer
        Number of digits after the decimal point for the results
    Returns
    ----------
    hpd: array with the lower 
          
    """
    sample = np.asarray(sample)
    sample = sample[~np.isnan(sample)]
    # get upper and lower bounds
    l = np.min(sample)
    u = np.max(sample)
    density = kde.gaussian_kde(sample)
    x = np.linspace(l, u, 2000)
    y = density.evaluate(x)
    #y = density.evaluate(x, l, u) waitting for PR to be accepted
    xy_zipped = zip(x, y/np.sum(y))
    xy = sorted(xy_zipped, key=lambda x: x[1], reverse=True)
    xy_cum_sum = 0
    hdv = []
    for val in xy:
        xy_cum_sum += val[1]
        hdv.append(val[0])
        if xy_cum_sum >= (1-alpha):
            break
    hdv.sort()
    diff = (u-l)/20  # differences of 5%
    hpd = []
    hpd.append(round(min(hdv), roundto))
    for i in range(1, len(hdv)):
        if hdv[i]-hdv[i-1] >= diff:
            hpd.append(round(hdv[i-1], roundto))
            hpd.append(round(hdv[i], roundto))
    hpd.append(round(max(hdv), roundto))
    ite = iter(hpd)
    hpd = list(zip(ite, ite))
    modes = []
    for value in hpd:
        x_hpd = x[(x > value[0]) & (x < value[1])]
        y_hpd = y[(x > value[0]) & (x < value[1])]
        modes.append(round(x_hpd[np.argmax(y_hpd)], roundto))
    return hpd, x, y, modes
```

```python
fit = az.from_cmdstan(["stan_output_%d_raw_%s_abundance_top%d_%s.csv"%(i,abundance_type,len(bacteria),taxonomy_level) for i in np.arange(1,4)])

lines = []
    
# basal growth rate
for taxa in bacteria:
    var = 'alpha__%s'%(taxa)
    data = []
    for i in np.arange(0,3):
        data.extend(list(fit.posterior[var][i].values))
    hpd_mu, x_mu, y_mu, modes_mu = hpd_grid(data)
    assert len(hpd_mu) == 1
    (x0, x1) = hpd_mu[0]
    lines.append(['basal_growth_rate', taxa, var, x0, x1, (x0+x1)/2, np.median(data), np.mean(data)/np.std(data), x0*x1>0])
    
# inulin response
for taxa in bacteria:
    var = 'epsilon__%s'%(taxa)
    data = []
    for i in np.arange(0,3):
        data.extend(list(fit.posterior[var][i].values))
    hpd_mu, x_mu, y_mu, modes_mu = hpd_grid(data)
    assert len(hpd_mu) == 1
    (x0, x1) = hpd_mu[0]
    lines.append(['diet_response', taxa, var, x0, x1, (x0+x1)/2, np.median(data), np.mean(data)/np.std(data), x0*x1>0])

# pairwise interactions
for taxa1 in bacteria:
    for taxa2 in bacteria:
        var = 'beta__%s_%s'%(taxa1,taxa2)
        data = []
        for i in np.arange(0,3):
            data.extend(list(fit.posterior[var][i].values))
        hpd_mu, x_mu, y_mu, modes_mu = hpd_grid(data)
        assert len(hpd_mu) == 1
        (x0, x1) = hpd_mu[0]
        lines.append(['pairwise_interaction', (taxa1,taxa2), var, x0, x1, (x0+x1)/2, np.median(data), np.mean(data)/np.std(data), x0*x1>0])
    
df_stan_output_summary = pd.DataFrame(lines, columns = ['Type','Taxa','Variable','Left','Right','Middle','Median','SNR','Significant'])
df_stan_output_summary.to_excel('stan_output_summary_%s_abundance_top%d_%s.xlsx'%(abundance_type,len(bacteria),taxonomy_level))
df_stan_output_summary.head()
```

# Plot contribution of five inulin responders to alpha diversity

```python
inulin_responders = ['Bacteroides-acidifaciens','Muribaculaceae','Faecalibaculum','Parasutterella','Bacteroides']
df_meta = pd.read_csv('../../../our_data/meta_data.csv', index_col=0)
df_meta = df_meta[df_meta.Diet=='Inulin']
df_bac = pd.read_csv('../../../our_data/16S_relative_abundance_species.csv', index_col=0)
df_meta_bac = pd.merge(df_meta, df_bac, left_index=True, right_index=True, how='inner')

lines = []
for sample in df_meta_bac.index:
    tmp = [sample]
    for responder in inulin_responders:
        prob = df_meta_bac.loc[sample,responder]
        if prob>0:
            tmp.append(-prob*np.log2(prob))
        else:
            tmp.append(0)
    others = 0
    total = 0
    for taxa in df_meta_bac.iloc[:,5:].columns:
        prob = df_meta_bac.loc[sample,taxa]
        if prob>0:
            total += (-prob*np.log2(prob))
            if taxa not in inulin_responders:
                others += (-prob*np.log2(prob))
    tmp.append(others)
    tmp.append(total)
    lines.append(tmp)
df_ad = pd.DataFrame(lines, columns=['SampleID']+inulin_responders+['Others','Total']).set_index('SampleID')
df_ad = pd.merge(df_meta_bac.iloc[:,0:5], df_ad,  left_index=True, right_index=True, how='inner')

fig, ax = plt.subplots(figsize=(12,2), nrows=1, ncols=4)
for k,vendor in enumerate(['Beijing','Guangdong','Hunan','Shanghai']):
    df_tmp = df_ad.loc[df_ad.Vendor==vendor,['Day']+inulin_responders+['Others','Total']].set_index('Day').unstack().reset_index()
    df_tmp.columns = ['Taxa','Day','SD']
    sns.lineplot(x='Day',y='SD', marker='o', hue='Taxa', ci=68, data=df_tmp, ax=ax[k])
    ax[k].set_title(vendor)
    ax[k].set_xlim([-1,32])
    if k!=3:
        ax[k].legend().remove()

plt.rcParams['svg.fonttype'] = 'none'
fig.savefig("contribution_inulin_responders_to_diversity.svg", format="svg")
```

# Compare data and simulations

```python
def gLV(t, y, bacterial_taxa, params):
    
    # check if y<0
    ypos = [m if m>0 else 0 for m in deepcopy(y)]
    
    deriv = []
    for k1,c1 in enumerate(bacteria):
        growth_rate_var = 'alpha__%s'%(c1)
        diet_response_var = 'epsilon__%s'%(c1)
        dc1_dt = params[growth_rate_var]+params[diet_response_var]
        for k2,c2 in enumerate(bacteria):
            pairwise_interaction_var = 'beta__%s_%s'%(c1,c2)
            dc1_dt += params[pairwise_interaction_var]*ypos[k2]
        deriv.append(dc1_dt*ypos[k1])
    return deriv
 
df_stan_output_summary = pd.read_excel('stan_output_summary_absolute_abundance_top20_species.xlsx', index_col=0)
params = dict(zip(df_stan_output_summary.Variable,df_stan_output_summary.Median))
df_bac_len(bacteria)_w_meta = df_bac_len(bacteria)_w_meta[df_bac_len(bacteria)_w_meta.Diet==1] # just inulin

# solve the model
all_mice = set(df_bac_len(bacteria)_w_meta.MiceID)
xcorr = np.linspace(0,31,311)
df_obs_pred = None
for curr_mice in all_mice:
    curr_df = df_bac_len(bacteria)_w_meta[df_bac_len(bacteria)_w_meta.MiceID==curr_mice]
    init_conc = list(curr_df.loc[curr_df.Day==0,bacteria].values)[0]
    # sol.t, sol.y (# number of taxa x numberr of time point)
    sol = solve_ivp(gLV, [0,31], init_conc, args=(bacteria,params), method='BDF', t_eval=xcorr)
    
    # HI4 is unstable in simulation and solver returns simulated value for only partial time points
    if np.abs(np.max(sol.t)-xcorr[-1])<1e-6:
        # get a copy of curr_df but replace observed values with simulated values
        curr_df_pred = deepcopy(curr_df)
        for k,t in enumerate(bacteria):
            cs = CubicSpline(sol.t, sol.y[k])
            curr_df_pred[t] = cs(curr_df_pred.Day)

        # add column to indicate observation or prediction
        curr_df['ValueType'] = 'Observation'
        curr_df_pred['ValueType'] = 'Prediction'
    
        if df_obs_pred is None:
            df_obs_pred = pd.concat([curr_df, curr_df_pred], ignore_index=True)
        else:
            df_obs_pred = pd.concat([df_obs_pred, curr_df, curr_df_pred], ignore_index=True)

# plot
fig, ax = plt.subplots(nrows=20, ncols=4, figsize=(15,40), sharex=True)
for j,vendor in enumerate(['Beijing','Guangdong','Hunan','Shanghai']):
    for i,t in enumerate(bacteria):
        _ = sns.lineplot(x='Day', y=t, hue='ValueType', data=df_obs_pred[df_obs_pred.Vendor==vendor], ax=ax[i,j])
        if i!=0 or j!=0:
            ax[i,j].legend().remove()
        ax[i,j].set_ylabel(t)
        if i==0:
            ax[i,j].set_title(vendor)
        
        if j==0:
            ax[i,j].set_ylabel(t.replace('_slash_','/').replace('_space_',' ').replace('_leftsquarebracket_','[').replace('_rightsquarebracket_',']').replace('_dash_','-').replace('_dot_','.').replace('_leftroundbracket','(').replace('_rightroundbracket_',')'))
        else:
            ax[i,j].set_ylabel('')
plt.tight_layout()
```

# Simulate effects of removing B. acidifaciens and Muribaculaceae

```python
def gLV(t, y, bacterial_taxa, params):
    
    # check if y<0
    ypos = [m if m>0 else 0 for m in deepcopy(y)]
    
    deriv = []
    for k1,c1 in enumerate(bacteria):
        growth_rate_var = 'alpha__%s'%(c1)
        diet_response_var = 'epsilon__%s'%(c1)
        dc1_dt = params[growth_rate_var]+params[diet_response_var]
        for k2,c2 in enumerate(bacteria):
            pairwise_interaction_var = 'beta__%s_%s'%(c1,c2)
            dc1_dt += params[pairwise_interaction_var]*ypos[k2]
        deriv.append(dc1_dt*ypos[k1])
    return deriv
 
df_stan_output_summary = pd.read_excel('stan_output_summary_absolute_abundance_top20_species.xlsx', index_col=0)
params = dict(zip(df_stan_output_summary.Variable,df_stan_output_summary.Median))
df_bac_len(bacteria)_w_meta = df_bac_len(bacteria)_w_meta[df_bac_len(bacteria)_w_meta.Diet==1] # just inulin

# solve the model
all_mice = df_bac_len(bacteria)_w_meta[['MiceID','Vendor']].drop_duplicates().sort_values(by='Vendor').MiceID
xcorr = np.linspace(0,5,300)
lines = []
for curr_mice in all_mice:
    curr_df = df_bac_len(bacteria)_w_meta[df_bac_len(bacteria)_w_meta.MiceID==curr_mice]
    vendor = list(set(curr_df['Vendor']))[0]
    
    # intact baseline
    init_conc1 = list(curr_df.loc[curr_df.Day==0,bacteria].values)[0]
    sol1 = solve_ivp(gLV, [0,5], init_conc1, args=(bacteria,params), method='BDF', t_eval=xcorr)
    
    # perturbed baseline
    init_conc2 = list(curr_df.loc[curr_df.Day==0,bacteria].values)[0]
    init_conc2[bacteria.index('Muribaculaceae')] = 0
    init_conc2[bacteria.index('Bacteroides_dash_acidifaciens')] = 0
    sol2 = solve_ivp(gLV, [0,5], init_conc2, args=(bacteria,params), method='BDF', t_eval=xcorr)
    
    #print(curr_mice, vendor, sol1.t[-1], sol2.t[-1])
    # HI4 is unstable in simulation and solver returns simulated value for only partial time points
    if np.abs(np.max(sol1.t)-xcorr[-1])<1e-6 and np.abs(np.max(sol2.t)-xcorr[-1])<1e-6:
        sol1_total = sol1.t*0
        sol2_total = sol2.t*0
        for k,taxa in enumerate(bacteria):
            sol1_total += sol1.y[k]
            sol2_total += sol2.y[k]
        density_ratio = sol2_total/sol1_total
        for k,v in zip(sol1.t, density_ratio):
            lines.append([curr_mice, vendor, k, v])
        
df_remove_Ba_Mu = pd.DataFrame(lines, columns=['MiceID','Vendor','Time','FC'])

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6,4))
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']
for k,vendor in enumerate(['Beijing','Guangdong','Hunan','Shanghai']):
    df_tmp = df_remove_Ba_Mu[df_remove_Ba_Mu.Vendor==vendor]
    for mice in set(df_tmp.MiceID):
        df_tmp2 = df_tmp[df_tmp.MiceID==mice]
        _ = sns.lineplot(x='Time', y='FC', data=df_tmp2, color=colors[k], ax=ax, label=vendor)
plt.tight_layout()

plt.rcParams['svg.fonttype'] = 'none'
fig.savefig("simulated_absolute_abundance_fold_change_remove_Ba_Mu.svg", format="svg")
```

# Simulate effects of removing positive outgoing links of B. acidaficiens and Muribaculaceae

```python
def gLV(t, y, bacterial_taxa, params, flag):
    
    # check if y<0
    ypos = [m if m>0 else 0 for m in deepcopy(y)]
    
    deriv = []
    for k1,c1 in enumerate(bacteria):
        growth_rate_var = 'alpha__%s'%(c1)
        diet_response_var = 'epsilon__%s'%(c1)
        dc1_dt = params[growth_rate_var]+params[diet_response_var]
        for k2,c2 in enumerate(bacteria):
            pairwise_interaction_var = 'beta__%s_%s'%(c1,c2)
            coef = params[pairwise_interaction_var]
            
            # only remove positive links from Ba or Mu to other taxa
            if flag and (c2=='Muribaculaceae' or c2=='Bacteroides_dash_acidifaciens') and coef>0:
                pass
            else:
                dc1_dt += coef*ypos[k2]
        deriv.append(dc1_dt*ypos[k1])
    return deriv
 
df_stan_output_summary = pd.read_excel('stan_output_summary_absolute_abundance_top20_species.xlsx', index_col=0)
params = dict(zip(df_stan_output_summary.Variable,df_stan_output_summary.Median))
df_bac_len(bacteria)_w_meta = df_bac_len(bacteria)_w_meta[df_bac_len(bacteria)_w_meta.Diet==1] # just inulin

# solve the model
all_mice = set(df_bac_len(bacteria)_w_meta.MiceID)
xcorr = np.linspace(0,5,100)
lines = []
for curr_mice in all_mice:
    curr_df = df_bac_len(bacteria)_w_meta[df_bac_len(bacteria)_w_meta.MiceID==curr_mice]
    vendor = list(set(curr_df['Vendor']))[0]
    init_conc = list(curr_df.loc[curr_df.Day==0,bacteria].values)[0]

    # intact interaction network
    sol1 = solve_ivp(gLV, [0,5], init_conc, args=(bacteria,params,0), method='BDF', t_eval=xcorr)
    
    # perturbed interaction network
    sol2 = solve_ivp(gLV, [0,5], init_conc, args=(bacteria,params,1), method='BDF', t_eval=xcorr)
    
    # print(curr_mice, vendor, sol1.t[-1], sol2.t[-1])
    # HI4 is unstable in simulation and solver returns simulated value for only partial time points
    if np.abs(np.max(sol1.t)-xcorr[-1])<1e-6 and np.abs(np.max(sol2.t)-xcorr[-1])<1e-6:
        sol1_total = sol1.t*0
        sol2_total = sol2.t*0
        for k,taxa in enumerate(bacteria):
            sol1_total += sol1.y[k]
            sol2_total += sol2.y[k]
        density_ratio = sol2_total/sol1_total
        for k,v in zip(sol1.t, density_ratio):
            lines.append([curr_mice, vendor, k, v])
        
df_remove_Ba_Mu = pd.DataFrame(lines, columns=['MiceID','Vendor','Time','FC'])

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6,4))
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']
for k,vendor in enumerate(['Beijing','Guangdong','Hunan','Shanghai']):
    df_tmp = df_remove_Ba_Mu[df_remove_Ba_Mu.Vendor==vendor]
    for mice in set(df_tmp.MiceID):
        df_tmp2 = df_tmp[df_tmp.MiceID==mice]
        _ = sns.lineplot(x='Time', y='FC', data=df_tmp2, color=colors[k], ax=ax, label=vendor)
ax.set_ylim([-0.03,1.05])
plt.tight_layout()

plt.rcParams['svg.fonttype'] = 'none'
fig.savefig("simulated_absolute_abundance_fold_change_remove_Ba_Mu.svg", format="svg")
```
