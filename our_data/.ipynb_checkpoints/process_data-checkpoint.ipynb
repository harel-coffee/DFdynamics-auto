```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_scfa = pd.read_csv('SCFA_BJ_Hongbin_1121.tsv', index_col=0, sep='\t')
df_scfa = df_scfa.drop('#q2:types')
df_scfa = df_scfa[df_scfa.Vendor=='Beijing']
df_scfa = df_scfa[df_scfa.Sample_location=='Stool']
df_scfa = df_scfa[['Vendor','Subject','Group','Day','Acetate','Butyrate','Propionate','Acetate_BJ','Butyrate_BJ','Propionate_BJ','Iso-butyrate_BJ','Valerate_BJ','Iso-valerate_BJ','Total_scfa_BJ']]
#df_scfa = df_scfa.dropna()
df_scfa[['Day','Acetate','Butyrate','Propionate','Acetate_BJ','Butyrate_BJ','Propionate_BJ','Iso-butyrate_BJ','Valerate_BJ','Iso-valerate_BJ','Total_scfa_BJ']]=df_scfa[['Day','Acetate','Butyrate','Propionate','Acetate_BJ','Butyrate_BJ','Propionate_BJ','Iso-butyrate_BJ','Valerate_BJ','Iso-valerate_BJ','Total_scfa_BJ']].astype(float)
df_scfa
```

```python
fig,ax = plt.subplots(nrows=1, ncols=3, figsize=(15,4))
sns.regplot(x='Acetate', y='Acetate_BJ', data=df_scfa, ax=ax[0])
sns.regplot(x='Butyrate', y='Butyrate_BJ', data=df_scfa, ax=ax[1])
sns.regplot(x='Propionate', y='Propionate_BJ', data=df_scfa, ax=ax[2])
```

```python
fig,ax = plt.subplots(nrows=1, ncols=3, figsize=(15,4))
sns.lineplot(x='Day', y='Acetate_BJ', hue='Group', data=df_scfa, ax=ax[0])
sns.lineplot(x='Day', y='Butyrate_BJ', hue='Group', data=df_scfa, ax=ax[1])
sns.lineplot(x='Day', y='Propionate_BJ', hue='Group', data=df_scfa, ax=ax[2])
```

```python
df_scfa_to_append = df_scfa[['Acetate_BJ','Butyrate_BJ','Propionate_BJ','Iso-butyrate_BJ','Valerate_BJ','Iso-valerate_BJ','Total_scfa_BJ']]
df_scfa_to_append.columns = ['Acetate','Butyrate','Propionate','Iso-butyrate','Valeric','Iso-valeric','Total']
df_scfa_to_append['Formate'] = np.nan
df_scfa_to_append['4-methylvaleric'] = np.nan
df_scfa_to_append['hexanoic'] = np.nan
df_scfa_to_append['heptylic'] = np.nan
df_scfa_to_append = df_scfa_to_append[['Formate','Acetate','Propionate','Iso-butyrate','Butyrate','Iso-valeric','Valeric','4-methylvaleric','hexanoic','heptylic','Total']]
df_scfa_to_append
```

```python
df_scfa_old = pd.read_excel('SCFA_old.xlsx', index_col=0)
df_scfa_new = df_scfa_old
df_scfa_new = df_scfa_new.loc[set(df_scfa_new.index)-set(df_scfa_to_append.index)]
df_scfa_new = pd.concat([df_scfa_new, df_scfa_to_append])
df_scfa_new.index.name = 'SampleID'
df_scfa_new = df_scfa_new.dropna(how='all')
df_scfa_new.head()
```

```python
df_scfa_new.to_excel('SCFA.xlsx')
```
