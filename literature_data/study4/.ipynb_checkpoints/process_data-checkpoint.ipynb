```python
import pandas as pd
import numpy as np
```

```python
df_tax = pd.read_csv('count.csv')
df_tax = df_tax[df_tax['rank']=='S']
df_tax = df_tax.drop('rank', axis=1).set_index('taxa', drop=True)
df_tax = df_tax.div(df_tax.sum(axis=0), axis=1)
df_tax = df_tax[(df_tax.T != 0).any()]
df_tax_T = df_tax.T
df_tax_T.index.name = 'SampleID'
df_tax_T.head()
```

```python
df_tax_T.to_excel('MGS_relative_abundance.xlsx')
```
