```python
import pandas as pd
import numpy as np
```

# Read taxonomy table

```python
df_tax = pd.read_excel('fos_metadata.xlsx', sheet_name='otutable')
lowest_classified_taxon = []
for idx in df_tax.index:
    taxon = df_tax.loc[idx,'#OTU ID']
    taxon = taxon.replace('k__','').replace('p__','').replace('o__','').replace('c__','').replace('f__','').replace('g__','').replace('s__','')
    taxon = [t.strip() for t in taxon.split(';')]
    taxon = [t if t!='' else 'unclassified' for t in taxon]
    for i in np.arange(len(taxon)):
        if taxon[i] == 'unclassified':
            if i==0:
                lowest_classified_taxon.append('unclassified')
            else:
                lowest_classified_taxon.append(taxon[i-1])
            break
        else:
            if i==len(taxon)-1:
                lowest_classified_taxon.append(taxon[i])

df_tax['#OTU ID'] = lowest_classified_taxon
df_tax = df_tax.groupby('#OTU ID').agg(sum)
for c in df_tax.columns:
    df_tax[c] = df_tax[c]/np.sum(df_tax[c])
    
df_tax_T = df_tax.T
df_tax_T.index.name = 'SampleID'
df_tax_T.to_excel('16S_relative_abundance.xlsx')

df_tax_T.head()
```
